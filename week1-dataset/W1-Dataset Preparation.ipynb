# 1 Image Recognition Task: Dataset Analysis & Visualization

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Helper function for dataset statistics
def dataset_info(X, y, dataset_name):
    print(f"Dataset: {dataset_name}")
    print(f"Shape of X: {X.shape}")
    print(f"Shape of y: {y.shape}")
    print(f"Number of classes: {len(np.unique(y))}")
    print(f"Classes: {np.unique(y)}")
    print("-" * 50)

# Visualization function
def visualize_samples(X, y, class_names, num_samples=5, cmap=None):
    plt.figure(figsize=(12, 5))
    for i in range(num_samples):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(X[i], cmap=cmap)
        plt.title(f"Label: {class_names[y[i]]}")
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

## 1.1 MNIST : handwrite number (0-9, grey) dataset

```python
# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Dataset Info
dataset_info(X_train, y_train, "MNIST")

# Visualize some samples
mnist_class_names = [str(i) for i in range(10)]
visualize_samples(X_train, y_train, mnist_class_names, cmap='gray')
```

## 1.2 Fashion MNIST : fashion product (10 classes, grey) dataset

```python
# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Dataset Info
dataset_info(X_train, y_train, "Fashion MNIST")

# Visualize some samples
fmnist_class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", 
                      "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
visualize_samples(X_train, y_train, fmnist_class_names, cmap='gray')
```

## 1.3 CIFAR-10 : colorful images (10 classes, colorful ) dataset

```python
# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Reshape y_train and y_test (CIFAR labels are 2D)
y_train = y_train.flatten()
y_test = y_test.flatten()

# Dataset Info
dataset_info(X_train, y_train, "CIFAR-10")

# Visualize some samples
cifar10_class_names = ["Airplane", "Automobile", "Bird", "Cat", "Deer", 
                       "Dog", "Frog", "Horse", "Ship", "Truck"]
visualize_samples(X_train, y_train, cifar10_class_names)
```

## 1.4 Dataset Summary

```python
# Number of samples per class
def class_distribution(y, class_names, dataset_name):
    unique, counts = np.unique(y, return_counts=True)
    plt.bar(class_names, counts, color='skyblue')
    plt.title(f"Class Distribution in {dataset_name}")
    plt.xlabel("Classes")
    plt.ylabel("Number of Samples")
    plt.xticks(rotation=45)
    plt.show()

# Analyze Class Distribution
class_distribution(y_train, mnist_class_names, "MNIST")
class_distribution(y_train, fmnist_class_names, "Fashion MNIST")
class_distribution(y_train, cifar10_class_names, "CIFAR-10")
```
