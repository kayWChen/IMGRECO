```python
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import time
import matplotlib.pyplot as plt
```

```python
# Parameter Setting

IMG_SIZE = 128
BATCH_SIZE = 32
SHUFFLE_SIZE = 1000
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)
DATASET = "cats_vs_dogs" #replace with other dataset from Tensorflow
num_classes = 2
NUM_FEATURES = 8100
```

```python
class DataLoader(object):
    def __init__(self, image_size, batch_size):

        self.image_size = image_size
        self.batch_size = batch_size

        # Load the dataset
        (self.train_data_raw, self.validation_data_raw, self.test_data_raw), self.metadata = tfds.load(
            DATASET,
            split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'], # Split the 'train' split into 80%, 10%, and 10%
            with_info=True,
            as_supervised=True
        )

        # Get the number of train examples
        self.num_train_examples = self.metadata.splits['train'].num_examples*80/100
        self.get_label_name = self.metadata.features['label'].int2str

        # Pre-process data
        self._prepare_data()
        self._prepare_batches()

    # Resize all images to image_size x image_size
    def _prepare_data(self):
        self.train_data = self.train_data_raw.map(self._resize_sample)
        self.validation_data = self.validation_data_raw.map(self._resize_sample)
        self.test_data = self.test_data_raw.map(self._resize_sample)

    # Resize one image to image_size x image_size
    def _resize_sample(self, image, label):
        image = tf.cast(image, tf.float32)
        image = (image/127.5) - 1
        image = tf.image.resize(image, (self.image_size, self.image_size))
        return image, label

    def _prepare_batches(self):
        self.train_batches = self.train_data.shuffle(1000).batch(self.batch_size)
        self.validation_batches = self.validation_data.batch(self.batch_size)
        self.test_batches = self.test_data.batch(self.batch_size)

    # Get defined number of  not processed images
    def get_random_raw_images(self, num_of_images):
        random_train_raw_data = self.train_data_raw.shuffle(1000)
        return random_train_raw_data.take(num_of_images)

    # Get defined number of processed images
    def get_random_resized_images(self, num_of_images):
        random_train_data = self.train_data.shuffle(1000)
        return random_train_data.take(num_of_images)
```

```python
data_loader = DataLoader(IMG_SIZE, BATCH_SIZE)

# show image samples
plt.figure(figsize=(10, 8))
i = 0
for img, label in data_loader.get_random_resized_images(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(img)
    plt.title("{} - {}".format(data_loader.get_label_name(label), img.shape))
    plt.xticks([])
    plt.yticks([])
    i += 1
plt.tight_layout()
plt.show()
```

```python
from skimage.feature import hog
from skimage import exposure
from skimage.color import rgb2gray
import numpy as np
import cv2

def get_features(dataset, method,
                 sigma=None,
                 selection=False, 
                 num_features=NUM_FEATURES
                ):
    """
    Extract features from a set of images using SIFT, SURF, or HOG.

    Parameters:
        input_datastore (list): List of file paths to the images.
        method (str): Feature extraction method ('SIFT', 'SURF', or 'HOG').
        sigma (float, optional): Sigma value for SIFT feature detection.

    Returns:
        features_out (np.ndarray): Extracted features.
        features_out2 (np.ndarray): Feature scales and metrics.
    """
    size_ds = len(dataset)
    features_out = np.zeros((size_ds, num_features))
    num_feature_points = int(num_features / 2)

    for i, (img, _) in enumerate(dataset):
        # Convert image to grayscale
        img_gray = rgb2gray(img) if len(img.shape) == 3 else img

        # Feature extraction based on method
        if method == 'SIFT':
            sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=sigma if sigma else 0.04)
            keypoints, descriptors = sift.detectAndCompute((img_gray * 255).astype('uint8'), None)
        elif method == 'SURF':
            surf = cv2.xfeatures2d.SURF_create()
            keypoints, descriptors = surf.detectAndCompute((img_gray * 255).astype('uint8'), None)
        elif method == 'ORB':
            orb = cv2.ORB_create()
            keypoints, descriptors = orb.detectAndCompute((img_gray * 255).astype('uint8'), None)
        elif method == 'HOG':
            # Extract HOG features
            features, hog_image = hog(img_gray, block_norm='L2-Hys', pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)
            # Normalize HOG image
            hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
            # Use the flattened HOG features
            descriptors = features
        else:
            raise ValueError(f"Unknown method: {method}")

#         # Process keypoints or descriptors
#         if method != 'HOG' and len(keypoints) > 0:
#             keypoints = sorted(keypoints, key=lambda k: k.response, reverse=True)
#             strongest_keypoints = keypoints[:num_feature_points]
#             scales = [kp.size for kp in strongest_keypoints]
#             metrics = [kp.response for kp in strongest_keypoints]
#         else:
#             strongest_keypoints = []
#             scales = []
#             metrics = []
#             print("Warning - not enough feature points!")

#         # Fill features_out2 matrix with scales and metrics
#         diff = num_feature_points - len(strongest_keypoints)
#         features_out[i, :num_feature_points] = scales + [0] * diff
#         features_out[i, num_feature_points:] = metrics + [0] * diff

        # If using HOG, store HOG features
        if method == 'HOG':
            features_out[i, :len(descriptors)] = descriptors

    return features_out

hog_feature = get_features(data_loader.get_random_resized_images(2),"HOG") 
print(hog_feature)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# normalize features
def normalize_features(features):
    """
    Normalize the features using z-score normalization.

    Parameters:
        features (np.ndarray): Extracted features (e.g., scales and metrics).

    Returns:
        np.ndarray: Normalized features with mean 0 and std deviation 1.
    """
    scaler = StandardScaler()
    return scaler.fit_transform(features)
```

```python
def data_preparation(method, split=False):
    
    start_time_FE = time.time()
    # feature extraction
    features_train = get_features(data_loader.train_data, method)
    features_validation = get_features(data_loader.validation_data, method)
    features_test = get_features(data_loader.test_data, method)
    end_time_FE = time.time()
    print(f"Feature extraction time-{method}: {end_time_FE - start_time_FE} seconds")

    # feature normalization
    features_train_normalized = normalize_features(features_train)
    features_validation_normalized = normalize_features(features_validation)
    features_test_normalized = normalize_features(features_test)

    # get labels
    labels_train = np.array([label.numpy() for _, label in data_loader.train_data])
    labels_validation = np.array([label.numpy() for _, label in data_loader.validation_data])
    labels_test = np.array([label.numpy() for _, label in data_loader.test_data])

    # prepare data(non-cross-validation)
    X_train = features_train_normalized
    y_train = labels_train

    X_validation = features_validation_normalized
    y_validation = labels_validation

    X_test = features_test_normalized
    y_test = labels_test
    
    # one-hot encoding for label     
    y_train = tf.keras.utils.to_categorical(y_train, num_classes)
    y_test = tf.keras.utils.to_categorical(y_test, num_classes)
    y_validation = tf.keras.utils.to_categorical(y_validation, num_classes)
    
    if not split:
        # concaten data for cross validation
        X = np.vstack((X_train, X_test, X_validation))
        y = np.hstack((y_train, y_test, y_validation))
        return X, y
    else:
        return X_train, y_train, X_validation, y_validation, X_test, y_test
```

```python
# neurual network classifier with full connection layer and output lays only
def cnn_model():
    
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(NUM_FEATURES,)),
        tf.keras.layers.Dense(64, activation='relu'),  # full connection layer
        tf.keras.layers.Dense(32, activation='relu'),  # full connection layer
        tf.keras.layers.Dense(num_classes, activation='softmax')  # output layer
    ])

    # model compiling
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

## SIFT feature + NN

```python
X_train, y_train, X_validation, y_validation, X_test, y_test = data_preparation("HOG",True)
```

```python
print(X_train.size, y_train.size)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model = cnn_model()

early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

start_time = time.time()

history = model.fit(X_train, y_train,
                    validation_data=(X_validation, y_validation),
                    epochs=500,
                    batch_size=64,
                    callbacks=[early_stopping])
                   

end_time = time.time()
# Record the training time
print(f"Training time: {end_time - start_time:.2f} seconds")
```

```python
# evaluation
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")

# training visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
