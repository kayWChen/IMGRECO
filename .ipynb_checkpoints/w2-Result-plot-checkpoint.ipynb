```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the histogram
categories = ["SIFT", "SURF", "ORB"]
feature_extraction_time = [271.14, 218.98, 161.98]
training_time = [4053.86, 6951.87, 3788.50]
best_accuracy = [0.5304, 0.5099, 0.5215]

# X positions for the groups
x = np.arange(len(categories))

# Width of the bars
width = 0.25

# Create the bar chart with colored labels
fig, ax1 = plt.subplots(figsize=(12, 6))

# Feature extraction time
bar1 = ax1.bar(x - width, feature_extraction_time, width, label="Feature Extraction Time (s)", color="steelblue")

# Training time
bar2 = ax1.bar(x, training_time, width, label="Training Time (s)", color="orange")

# Create secondary y-axis for accuracy
ax2 = ax1.twinx()

# Best accuracy
bar3 = ax2.bar(x + width, best_accuracy, width, label="Best Accuracy", color="green")

# Set labels and titles
ax1.set_xlabel("Feature Extraction Method", fontsize=12)
ax1.set_ylabel("Time (s)", fontsize=12)
ax2.set_ylabel("Accuracy", fontsize=12)
ax1.set_xticks(x)
ax1.set_xticklabels(categories)
plt.title("Comparison of Feature Extraction Methods", fontsize=14)

# Combine legends from both axes with matching colors
legend_elements = [
    plt.Line2D([0], [0], color="steelblue", lw=4, label="Feature Extraction Time (s)"),
    plt.Line2D([0], [0], color="orange", lw=4, label="Training Time (s)"),
    plt.Line2D([0], [0], color="green", lw=4, label="Best Accuracy")
]
ax1.legend(handles=legend_elements, loc="upper left", fontsize=9)

# Show the plot
plt.tight_layout()
plt.show()
```
